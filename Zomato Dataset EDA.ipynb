```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating counting'})
```

```python
ratings
```

## Observation
1. When the Rating is between 4.5 to 4.9 -------> Excellent
2. When the Rating is between 4.0 to 4.4 -------> Very Good
3. When the Rating is between 3.5 to 3.9 -------> Good
4. When the Rating is between 3.0 to 3.4 -------> Average
5. When the Rating is between 2.5 to 2.9 -------> Average
6. When the Rating is less than 2.4      -------> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating counting",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating counting",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

Observation:
      1. Not Rated count is very high.
      2. Maximum number of ratings are between 2.5 to 3.4
      3. The highest rating count is recorded between 3.0 to 3.1
      4. The lowest rating count is recorded between 1.8 to 1.9

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','purple'])
```

Observation :- It is how frequently the ratings are present or occuring
